```python
with open('boblist2.txt') as file:
    f = file.read()
    titleblock = re.split("\n",f)
    TitleConceptMap = {}
    for i in range(len(titleblock)):
        block = titleblock[i].strip()
        if re.search("<title/",block):
            title = re.split(">",block)[1]
            title = re.split("<",title)[0].strip()
            terms = titleblock[i+1]
            terms = re.split(">",terms)[1]
            terms = re.split("<",terms)[0]
            terms = re.split(",",terms)
            TitleConceptMap[title] = terms
            
            
```

```python
termNet = []
for title in TitleConceptMap:
    for term in TitleConceptMap[title]:
        if (term not in termNet):
            termNet.append(term)
            
        
```

```python
termIntersections= {};
for term in termNet:
    termIntersections[term] = [];
    for title in TitleConceptMap:
        if (term in TitleConceptMap[title]):
            termIntersections[term].append(title);
print(termIntersections)
```

```python
for term in termIntersections:
    termIntersections[term] = set(termIntersections[term]);
T = nx.Graph()
for term1 in termNet:
    for term2 in termNet:
        if (term1 != term2 and term1 !='' and term2 != ''):
            intersection = termIntersections[term1].intersection(termIntersections[term2]);
            if (len(intersection)>0):
                T.add_edge(term1,term2, weight = len(intersection))
nx.write_weighted_edgelist(T, 'bob2TermNet.edgelist')
nx.write_gml(T,'bob2TermNet.gml')
```
